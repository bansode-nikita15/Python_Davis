**1.You are given a NumPy array of student marks. Write a program to extract marks greater than the class average.**

```python
import numpy as np
stud_marks=np.array([10,54,67,47,76,89,54,56])
class_avg=np.mean(stud_marks)
result=stud_marks[stud_marks>class_avg]
print("class average:",class_avg)
print("marks greater than class average:",result)
```

***10.Given a NumPy array of sensor readings, write a program to detect and extract all abnormal readings that are more than 2 times the mean value.***




```python
import numpy as np
readings = np.array([10, 12, 11, 13, 50, 14, 12])
mean_val = np.mean(readings)
abnormal = readings[readings > 2 * mean_val]
print("Abnormal readings:", abnormal)
```

***11.A NumPy array contains salary data of employees. Write a program to apply tax rules:***
Salary above 50,000 → deduct 10 percent
Salary above 1,00,000 → deduct 20 percent


```python
salary = np.array([30000, 60000, 120000, 80000],dtype=float)
final_salary = salary.copy()
final_salary[salary > 100000] *= 0.8
final_salary[(salary > 50000) & (salary <= 100000)] *= 0.9
print("Salary after tax:", final_salary)
```

***12.Given a 2D NumPy array of student marks, write a program to find how many students failed in each subject.***

***14.Given a NumPy array of daily profit and loss values, write a program to separate profit days and loss days using boolean indexing.***

```python
profit_loss = np.array([500, -200, 300, -100, 0, 400, -50])
profit_days = profit_loss[profit_loss > 0]
loss_days = profit_loss[profit_loss < 0]
print("Profit days:", profit_days)
print("Loss days:", loss_days)
```

***15.A 2D NumPy array represents attendance of employees where 1 means present and 0 means absent. Write a program to calculate attendance percentage of each employee.***

```python
attendance = np.array([
    [1, 1, 0, 1, 1],
    [1, 0, 0, 1, 0],
    [1, 1, 1, 1, 1]
])
attendance_percentage = np.mean(attendance, axis=1) * 100
print("Attendance percentage:", attendance_percentage)
```

***16.A NumPy array contains ratings given by users. Write a program to normalize the ratings between 0 and 1.***




```python
ratings = np.array([1, 3, 4, 5, 2])
normalized = (ratings - np.min(ratings)) / (np.max(ratings) - np.min(ratings))
print("Normalized ratings:", normalized)
```

***17.Given a NumPy array of transaction amounts, write a program to flag suspicious transactions that are either extremely low or extremely high.***

```python
transactions = np.array([200, 1500, 60000, 450, 30000, 100000])
suspicious = transactions[(transactions < 500) | (transactions > 50000)]
print("Suspicious transactions:", suspicious)
```

***18.A 2D NumPy array stores marks of students across tests. Write a program to identify the topper based on total marks***

```python
marks = np.array([
    [80, 90, 85],
    [70, 75, 72],
    [95, 88, 92]
])
total_marks = np.sum(marks, axis=1)
topper = np.argmax(total_marks)
print("Total marks:", total_marks)
print("Topper student index:", topper)
```
