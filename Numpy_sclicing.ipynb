## ***2D sclicing Numpy***

```python
arr=np.array([
    [10,20,30,53],
    [40,50,60,34],
    [70,80,90,56]
])
print(arr[1:4,0:5])
```

```python
#axis=0 /column
#axis=1/ row
marks=np.array([
    [10,20,30],
    [40,50,60],
    [70,80,90]
])
print("Average Marks",np.mean(marks,axis=1))
```

```python
print("Average Marks",np.mean(marks,axis=0))
```
